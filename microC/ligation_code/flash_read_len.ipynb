```python
from Bio import SeqIO
import matplotlib.pyplot as plt
from collections import Counter

def plot_seq_length_distribution(fastq_file,sample_name):
    # Read the sequence lengths from the FASTQ file
    seq_lengths = [len(record.seq) for record in SeqIO.parse(fastq_file, "fastq")]
    
    # Count the frequency of each length
    length_freq = Counter(seq_lengths)
    
    # Sort the lengths
    lengths = sorted(length_freq.keys())
    freqs = [length_freq[length] for length in lengths]
    
    # Find the length with the highest frequency
    max_freq_length = max(length_freq, key=length_freq.get)
    max_freq = length_freq[max_freq_length]
    
    # Plot the length distribution
    plt.plot(lengths, freqs)
    plt.xlabel('Sequence Length')
    plt.ylabel('Frequency')
    plt.title(f'{sample_name} Sequence Length Distribution')
    
    # Mark the length with the highest frequency
    plt.plot(max_freq_length, max_freq, 'ro')
    plt.annotate(f'Max length: {max_freq_length}bp', (max_freq_length, max_freq))
    
    plt.show()

# Usage
fastq_file = '/mnt/hpc/home/xuxinran/microC/date/date_0520/C_flash_map/more1k/C_L003_more1k-flash.extendedFrags.fastq'
plot_seq_length_distribution(fastq_file,"scMicroC C(>1k)")
```
